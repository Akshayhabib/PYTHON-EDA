# Zomato Dataset Exploratory Data Analysis

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
zomato_df=pd.read_csv('Zomatodataset/zomato.csv')
zomato_df.head()
```

```python
zomato_df.columns
```

```python
zomato_df = zomato_df.drop(['Longitude','Latitude'],axis=1)
```

```python
zomato_df.describe()
```

```python
[features for features in zomato_df.columns if zomato_df[features].isnull().sum()>0]
```

```python
country_df=pd.read_excel('Zomatodataset/Country-Code.xlsx')
country_df.head()
```

```python
final_df=pd.merge(zomato_df,country_df,on='Country Code',how='left')
final_df.head()
```

```python
final_df.columns
```

```python
#<TOP 5 COUNTRIES THAT USES ZOMATO
country_val=final_df['Country'].value_counts()
country_names=final_df['Country'].value_counts().index
plt.figure(figsize=(15,20))
plt.pie(country_val[:5],labels=country_names[:5],autopct='%1.1f%%')
plt.legend(country_names)
plt.show()
```

- As you can see maximum(93.2%) records are from India ,after that United States (4.7) and then United Kingdom (0.9%),Brazil & South Africa(0.6%)

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings.head()
```

- When Rating is between 4.5 to 4.9 --->>Excellent
- When Rating  isbetween 4.0 to 4.4 --->>very good
- when Rating is between 3.5 to 3.9 ---->>good
- when Rating is between 3.0 to 3.4 ---->>average
- when Rating is between 2.5 to 2.9 ---->>average
- when Rating is between 1.8 to 2.4 ---->>Poor

```python
plt.figure(figsize=(15,8))
sns.barplot(x='Aggregate rating',y='Rating_count',hue='Rating color',data=ratings, palette='inferno')
```

- Not Rated count is very high
- Maximum number of rating are between 2.5 to 3.4

```python
#count plot
#plt.figure(figsize=(15,8))
sns.countplot(x='Rating color',data=ratings, palette='inferno')
```
